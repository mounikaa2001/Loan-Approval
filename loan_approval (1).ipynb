#Description of columns

- Loan_ID: A unique identifier for each loan application.
- Gender: The gender of the applicant (Male/Female).
- Married: The marital status of the applicant (Yes/No).
- Dependents: The number of dependents the applicant has.
- Education: The applicant's educational background (Graduate/Not Graduate).
- Self_Employed: Whether the applicant is self-employed (Yes/No).
- ApplicantIncome: The applicant's annual income.
- CoapplicantIncome: The co-applicant's annual income (if applicable).
- LoanAmount:** The amount of loan requested.
- Loan_Amount_Term: The term of the loan in months.
- Credit_History: The applicant's credit history (1 indicates good credit, 0 indicates bad credit).
- Property_Area: The location of the property for which the loan is being sought (Urban/Semiurban/Rural).
- Loan_Status: Whether the loan was approved or not (Y/N).

## Detection of outliers using IQR

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_copy['ApplicantIncome'])
plt.title('Box Plot for Outlier Detection in ApplicantIncome')
plt.show()
```

```python
Q1 = df_copy['ApplicantIncome'].quantile(0.25)
Q3 = df_copy['ApplicantIncome'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_income = df_copy[(df_copy['ApplicantIncome'] < lower_bound) | (df_copy['ApplicantIncome'] > upper_bound)]
print('count of outliers in coapplicantincome :' ,len(outliers_income))
#print("Outliers in ApplicantIncome:\n", outliers_income)
```

```python
outliers_income.index
```

```python
df_copy.shape
```

```python
import numpy as np # Import numpy in case it wasn't imported previously

# Drop rows with outliers and get the indices of dropped rows
dropped_rows_indices = outliers_income.index
df_copy = df_copy.drop(dropped_rows_indices) # Use the .drop method on the DataFrame

# Check for missing values after dropping rows (this might not be necessary if you are sure there are no missing values)
print(df_copy['ApplicantIncome'].isnull().sum())

# If there are any missing values, fill them with the median
if df_copy['ApplicantIncome'].isnull().sum() > 0:
    df_copy['ApplicantIncome'].fillna(df_copy['ApplicantIncome'].median(), inplace=True)

# Visualize the distribution after handling outliers
import seaborn as sns # Import seaborn in case it wasn't imported previously
import matplotlib.pyplot as plt # Import matplotlib.pyplot in case it wasn't imported previously
sns.boxplot(data=df_copy['ApplicantIncome'])
plt.show()
```

# Summary of the Work

## Data Cleaning and Preprocessing:
 - Outliers in 'ApplicantIncome', 'CoapplicantIncome', and 'LoanAmount' were identified using box plots and IQR method and subsequently removed.
 - Missing values in these columns, if any, were filled with the median value.

## Exploratory Data Analysis (EDA):
 - Histograms were used to visualize the distribution of numerical variables ('ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term') by loan status ('Approved' vs. 'Rejected').
 - Count plots were used to visualize the frequency distribution of categorical variables ('Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area') by loan status.
 - Pie charts were used to represent the composition of categorical variables.

## Correlation Analysis:
 - A correlation matrix was computed to identify relationships between numerical variables.
 - A bar chart was plotted to visualize the correlation of 'Credit_History' with other numerical variables.
 - A heatmap was used to visualize the correlation matrix, highlighting positive and negative correlations.

## Key Insights:

 - Lower-income individuals and those with lower coapplicant incomes are more likely to apply for loans.
 - Loan approval rates are generally higher for lower loan amounts.
 - Graduates and married individuals have a higher likelihood of loan approval.
 - Semi-urban areas show the highest loan application and approval rates.
 - There is a positive correlation between applicant income and loan amount, suggesting higher income applicants tend to apply for larger loans.
- Credit history shows a slight positive correlation with loan amount, indicating good credit history might lead to larger loan approvals.

## Potential Applications:

- The insights gained from this analysis can be used to tailor loan products and marketing strategies to target specific demographics.
- Risk assessment and approval criteria can be adjusted based on applicant characteristics.
- Areas for improvement in loan accessibility and approval processes can be identified.
